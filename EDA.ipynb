```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Data Gathering 

```python
file_path = 'data/c.csv'
```

```python
# Number of Columns per File (Ingredients plus ID)

# num_columns = 3 + 1 # file a
# num_columns = 10 + 1 # file b
num_columns = 1010 + 1 # file c
# num_columns = 38 + 1 # file d
# num_columns = 38 + 1 # file e
custom_header = [str(i) for i in range(num_columns)]
```

```python
# Schema 
# string_dtype = 'str'    # Files A B
string_dtype = 'str'  # Files C D E

schema = {f'{i}': string_dtype for i in range(1, num_columns)}
schema['0'] = 'int'
```

```python
# Define your schema as a dictionary where keys are column names and values are data types
# Use the 'str' data type with a maximum string length

# Assuming the data in your CSV file is separated by spaces, you can specify the delimiter as ' '
delimiter = ' '

# Read the CSV file into a Pandas DataFrame, skipping the first line
df = pd.read_csv(file_path, delimiter=delimiter, skiprows=1, header=None, names=custom_header, dtype=schema)
```

```python
display(df)
```

## Display Ingrediants of Pizzas 

```python
# Count the number of non-null values in each row and add to a new column 'Count'
df['Count'] = df.count(axis=1)

# Display the DataFrame with the new 'Count' column
print(df)
```

```python
# Create a histogram of the 'Count' column
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.hist(df['Count'], rwidth=0.8, alpha=0.75, color='b')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.title('Histogram of Non-NA Value Counts')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
